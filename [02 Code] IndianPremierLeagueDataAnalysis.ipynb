# Multivariete Analysis 

## correlation matrix for numerical features by using Heat map

```python
# Compute the correlation matrix for numerical features
correlation_matrix = df.corr()

# Create a heat map to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heat Map of Numerical Features')
plt.tight_layout()

# Display the heat map
plt.show()
```

## How does the number of matches played in each season and city change over the years?

```python


# Create a pivot table to count the number of matches played in each season and city
matches_per_city_season = df.pivot_table(values='ID', index='City', columns='Season', aggfunc='count', fill_value=0)

# Create a heat map to visualize the change in the number of matches played in each season and city over the years
plt.figure(figsize=(12, 8))
sns.heatmap(matches_per_city_season, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)
plt.title('Number of Matches Played in Each Season and City')
plt.xlabel('Season')
plt.ylabel('City')
plt.tight_layout()

# Display the heat map
plt.show()
```

## How does the performance of teams vary in terms of the number of wickets lost and runs scored in different venues?


```python
# Create a scatter plot to visualize the number of wickets lost and runs scored for different teams at different venues
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WonBy', y='Margin', hue='WinningTeam', style='Venue', data=df, palette='Set1')
plt.title('Performance of Different Teams in Terms of Wickets Lost and Runs Scored at Different Venues')
plt.xlabel('Won By (Wickets)')
plt.ylabel('Winning Margin (Runs)')
plt.legend(title='Winning Team', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Display the scatter plot
plt.show()
```

## How does the performance of teams vary when they win the toss and bat first vs. bat second?

```python
matches_with_toss_decision = df[df['TossDecision'].isin(['bat', 'field'])]

# Create a box plot to visualize the distribution of winning margins by runs for different teams based on toss decision
plt.figure(figsize=(12, 6))
sns.boxplot(x='WinningTeam', y='Margin', hue='TossDecision', data=matches_with_toss_decision, palette='Set3')
plt.title('Winning Margins by Runs for Different Teams Based on Toss Decision')
plt.xlabel('Winning Team')
plt.ylabel('Winning Margin (Runs)')
plt.legend(title='Toss Decision', loc='upper right', labels=['Bat First', 'Bat Second'])
plt.xticks(rotation=45)
plt.tight_layout()

# Display the box plot
plt.show()
```

```python
## How does the toss decision (batting/fielding) affect the number of wickets lost and runs scored by different teams?
```

```python

# Create a scatter plot to visualize the number of wickets lost and runs scored for different teams based on toss decision
plt.figure(figsize=(10, 6))
sns.scatterplot(x='WonBy', y='Margin', hue='WinningTeam', style='TossDecision', data=df, palette='Paired')
plt.title('Impact of Toss Decision on Number of Wickets Lost and Runs Scored for Different Teams')
plt.xlabel('Won By (Wickets)')
plt.ylabel('Winning Margin (Runs)')
plt.legend(title='Winning Team', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Display the scatter plot
plt.show()
```

## How does the average margin of victories (by runs) change for matches won by different teams over the years?

```python
matches_won_by_runs = df[df['WonBy'] == 'Runs']

# Group data by season and team and calculate the average margin of victories
team_average_margin = matches_won_by_runs.groupby(['Season', 'WinningTeam'])['Margin'].mean().reset_index()

# Create a bar plot to visualize the average margin of victories by runs for different teams over the years
plt.figure(figsize=(12, 6))
sns.barplot(x='WinningTeam', y='Margin', hue='Season', data=team_average_margin, ci=None, palette='coolwarm')
plt.title('Average Margin of Victories by Runs for Different Teams over the Years')
plt.xlabel('Winning Team')
plt.ylabel('Average Margin of Victories (Runs)')
plt.legend(title='Season', loc='upper right', labels=df['Season'].unique())
plt.xticks(rotation=45)
plt.tight_layout()

# Display the bar plot
plt.show()
```

## How does the performance of the top player (by "Player of the Match" awards) change over the years

```python

# Filter data for the top player (by "Player of the Match" awards)
top_player = df['Player_of_Match'].value_counts().idxmax()
top_player_performance = df[df['Player_of_Match'] == top_player]

# Create a line plot to visualize the performance of the top player over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x='Season', y='Margin', data=top_player_performance, marker='o', color='b')
plt.title(f'Performance of {top_player} over the Years')
plt.xlabel('Season')
plt.ylabel('Winning Margin')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Display the line plot
plt.show()
```

```python
## How do the performances of teams change over the years in terms of the number of wins and the margin of victories
```

```python
# Group data by season and team and calculate the number of wins and average margin of victories
team_performance = df.groupby(['Season', 'WinningTeam']).agg({'ID': 'count', 'Margin': 'mean'}).reset_index()
team_performance.rename(columns={'ID': 'MatchesWon', 'Margin': 'AverageMargin'}, inplace=True)

# Create a scatter plot to visualize the number of wins and average margin of victories for each team over the years
plt.figure(figsize=(12, 6))
sns.scatterplot(x='MatchesWon', y='AverageMargin', hue='WinningTeam', style='Season', data=team_performance, palette='tab10')
plt.title('Team Performance over the Years: Number of Wins vs. Average Margin of Victories')
plt.xlabel('Number of Wins')
plt.ylabel('Average Margin of Victories')
plt.legend(title='Season', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Display the scatter plot
plt.show()
```

## How does the toss decision (batting/fielding) influence the match outcome (win/loss) for different teams in each season?(multivariete)

```python
# Create a count plot to visualize the relationship between toss decision, match outcome, and season for different teams
plt.figure(figsize=(12, 6))
sns.countplot(x='WinningTeam', hue='TossDecision', data=df, hue_order=['bat', 'field'], palette='Set2')
plt.title('Toss Decision vs. Match Outcome for Different Teams in Each Season')
plt.xlabel('Winning Team')
plt.ylabel('Count')
plt.legend(title='Toss Decision', loc='upper right', labels=['Bat', 'Field'])
plt.xticks(rotation=45)
plt.tight_layout()

# Display the count plot
plt.show()
```
